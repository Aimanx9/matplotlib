# Basic Line Plot

```python
daily_return_df = pd.read_csv('/content/drive/MyDrive/Matplotlib/daily_returns.csv')
daily_return_df
```

# Scatter Plot

```python

X = daily_return_df['AAPL']
Y = daily_return_df['sp500']
plt.scatter(X, Y)
plt.ylabel('Apple Daily Returns')
plt.xlabel('S&P500 Daily Returns');
```

# Pie Chart

```python
values = [20, 55, 5, 17, 4]
colors = ['g', 'r', 'y', 'b', 'm']
labels = ['AAPL', 'GOOG', 'T', 'TSLA', 'AMZN']

plt.figure(figsize=(7, 7))
plt.pie(values, colors=colors, labels = labels)
plt.title('Stock Portfolio')
```

```python
num_bins = 40
plt.figure(figsize=(7,5))
plt.hist(daily_return_df['GOOG'],num_bins);
plt.grid()
plt.title('Histogram: mu =' + str(mu) + ', sigma=' + str(sigma) );
```

# Multiple Plots

```python
stock_df.plot(x = 'Date', y= ['AAPL', 'sp500', 'GOOG'], linewidth = 3)
plt.ylabel('Price')
plt.title('Apple vs S&P 500')
plt.grid()
plt.legend(loc='upper center');
```

# Subplot

```python
plt.figure(figsize=(8,8))
plt.subplot(3, 1, 1)
plt.plot(stock_df['GOOG'], 'r--')
plt.grid()
plt.legend(['Google Price'])
plt.title('Google vs Apple vs S&P 500')

plt.subplot(3, 1, 2)
plt.plot(stock_df['AAPL'],'b--')
plt.grid()
plt.legend(['Apple Price'])

plt.subplot(3, 1, 3)
plt.plot(stock_df['sp500'],'g--')
plt.grid()
plt.legend(['S&P 500 Price']);
```

# 3D Plot

```python
# https://matplotlib.org/mpl_toolkits/index.html

from mpl_toolkits.mplot3d import Axes3D
fig =plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection = '3d')
#convert to list
x = daily_return_df['AAPL'].tolist()
y = daily_return_df['GOOG'].tolist()
z = daily_return_df['sp500'].tolist()

ax.scatter(x,y,z, c = 'r', marker = 'o')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_zlabel('z label');
```

# Boxplot

```python
#numpy.random.normal() takes three arguments: mean, standard deviation of the normal distribution, and number of values desired
#https://stackoverflow.com/questions/17725927/boxplots-in-matplotlib-markers-and-outliers
```

```python
np.random.seed(20)
data_1 = np.random.normal(200, 20, 2000)
data_2 = np.random.normal(60, 30, 2000)
data_3 = np.random.normal(70, 20, 2000)
data_4 = np.random.normal(40, 5, 2000)

data = [data_1, data_2, data_3, data_4]
fig = plt.figure(figsize = (10,7))
ax = fig.add_subplot(111)
bp = ax.boxplot(data)
```
